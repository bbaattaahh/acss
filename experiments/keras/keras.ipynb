```python
import keras
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from keras.models import Sequential
from keras.layers import Convolution2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
import numpy as np
import glob
import pandas as pd
import cv2
```

```python
# Dummy example
model = Sequential()
model.add(Dense(1, input_dim=784, activation='sigmoid'))
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# generate dummy data
import numpy as np
data = np.random.random((1000, 784))
labels = np.random.randint(2, size=(1000, 1))

# train the model, iterating on the data in batches
# of 32 samples
model.fit(data, labels, nb_epoch=10, batch_size=32)
```

```python
model = Sequential()
model.add(Dense(1, input_dim=7500, activation='sigmoid'))
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
#Example with own data
import glob
training_folder = 'c:\\Users\\Henrik\\Dropbox\\acss\\images\\heads\\'
```

```python
files_to_transform = glob.glob(training_folder + '*.jpg')
out_list = []
```

```python
files_to_transform[0]
```

```python
for f in files_to_transform:
        image = cv2.imread(f)

        if image.shape != (50,50,3):
            image_resized = cv2.resize(image, (50,50))
        else:
            image_resized = image

        flat_image = image_resized.reshape(1, 7500)
        y_hat = model.predict_classes(flat_image, verbose = 1)[0]
        name = f.split(training_folder)[1].split(".jpg")[0]
        row_content = [name, y_hat]
        out_list = out_list + [row_content]

        
df = pd.DataFrame(out_list)   
df.to_csv(path_or_buf = "prediction_neuro_classes.csv", sep = ";", index = False)
```
